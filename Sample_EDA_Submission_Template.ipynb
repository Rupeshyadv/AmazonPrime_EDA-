## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ===============================
# REMOVE DUPLICATES
# ===============================

credits_df_clean = credits_df.drop_duplicates().copy()
titles_df_clean = titles_df.drop_duplicates().copy()

# ===============================
# HANDLE MISSING VALUES (CREDITS)
# ===============================

# We DO NOT impute 'character' because:
# - Missing means the person is NOT an actor (director/writer)
# - Imputation would add wrong information
# → We leave it as is.


# ===============================
# HANDLE MISSING VALUES (TITLES)
# ===============================

# Impute description with a neutral text
titles_df_clean["description"] = titles_df_clean['description'].fillna("No Description Available")

# Impute age_certification with 'Not Rated'
titles_df_clean["age_certification"] = titles_df_clean['age_certification'].fillna("Not Rated")

# Seasons → Movies have season = 0 (correct representation)
titles_df_clean.loc[titles_df_clean['type'] == "MOVIE", "seasons"] = \
    titles_df_clean.loc[titles_df_clean['type'] == "MOVIE", "seasons"].fillna(0)

# NOTE:
# We are intentionally NOT imputing:
# - imdb_id
# - imdb_score
# - imdb_votes
# - tmdb_score
# - tmdb_popularity
# because missing values have meaning in ratings/IDs.


# ===============================
# SHOW FINAL INFO
# ===============================

print("\nCredits Dataset (Cleaned):")
credits_df_clean.info()

print("\nTitles Dataset (Cleaned):")
titles_df_clean.info()

# ===============================
# MERGE DATASETS FOR FURTHER ANALYSIS/VISUALIZATION
# ===============================

# Merge credits and titles datasets on 'id', which is common in both datasets
merged_df = pd.merge(titles_df_clean, credits_df_clean, on='id', how='left')
print("\nMerged Dataset Info:")
merged_df.info()
display(merged_df.head())
display(merged_df.tail())

# check if still have missing values after merge
print("\nMerged Dataset Missing Values:\n", merged_df.isnull().sum())

# NOTE:
# After merging titles and credits, some rows have missing values in 
# 'person_id', 'name', 'role', and 'character'.
#
# This happens because not every title in titles.csv has corresponding
# cast/crew information in credits.csv. The merge keeps all titles 
# (LEFT JOIN), so titles without credits naturally produce NaN values.
#
# These missing values should NOT be imputed because:
# - They do not represent actual missing information.
# - They simply indicate that cast/crew data was not provided for that title.
#
# We keep these NaN values as-is since imputing them would introduce 
# incorrect or artificial information into the dataset.
````

### What all manipulations have you done and insights you found?

### **Manipulations**

1. Removed duplicate rows from both datasets to ensure data integrity.
2. Handled missing values appropriately based on column semantics:
3. titles.csv: imputed description, age_certification, and assigned seasons = 0 for movies.
4. credits.csv: did not impute character because missing values naturally occur for non-actors.
5. Merged titles.csv and credits.csv using a LEFT JOIN on id to combine title information with cast and crew details.

````python
# Top 10 genres in the Dataset
# Goal -> Identify the top 10 most common genres in the catalog

# aggregate genre counts
import ast

def to_list_safe(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return x

# Convert strings → lists
# Here some enteries in 'genres' are strings representing lists (e.g. "['Drama', 'Action']") and some are actual lists.
# We need to convert the string representations into actual lists for proper processing.
merged_df['genres'] = merged_df['genres'].apply(to_list_safe)

# Explode
genre_exploded = merged_df['genres'].explode().str.strip()

# Count top 10
genre_counts = genre_exploded.value_counts().head(10).reset_index()


genre_counts.columns = ['genre', 'count']
print(genre_counts)

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(data=genre_counts, x='genre', y='count')
plt.xticks(rotation=45)
plt.title("Top 10 Genres")
plt.show()

````

##### 1. Why did you pick the specific chart?

I used bar plot because it clearly highlights which genres dominate the platform.

##### 2. What is/are the insight(s) found from the chart?

- Drama, Comedy, Action appear as top genres.
- Genres like Horror, european, scifi & fantasy have fewer titles.
- Shows diversity but also gaps.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The chart shows that some genres are under-represented, which may cause viewers who prefer those genres to switch to competitors. By expanding content in these low-volume genres, Amazon Prime can improve catalog balance, attract niche audiences, and increase overall user retention and revenue.

````python
# Age Certification vs IMDb score
# Goal -> Analyze how age certification relates to IMDb ratings

# Box plot 
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x='age_certification', y='imdb_score')
plt.title("IMDb Score Distribution by Age Certification")
plt.xlabel("Age Certification")
plt.ylabel("IMDb Score")
plt.show()
````

##### 1. Why did you pick the specific chart?

I used box plot because it is the most effective way to compare the distribution of IMDb scores across multiple age certification categories. Unlike bar charts or line charts, a box plot shows:

- Median score → the central tendency for each age group
- Spread (IQR) → how consistent or varied the scores are
- Whiskers → overall score range
- Outliers → extremely high- or low-rated titles

##### 2. What is/are the insight(s) found from the chart?

- TV-PG content has the highest median IMDb score, indicating that family-friendly titles generally receive better audience reception.
- TV-MA content shows the widest score distribution, meaning adult content is more unpredictable — it includes both top-performing hits and low-rated titles.
- Restricted-age categories (R, TV-MA) tend to have more variance, while PG content remains consistent.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The pattern indicates that family-friendly (TV-PG) content performs more consistently well, which means investing in high-quality PG-rated shows could strengthen Prime Video’s reliability and broader audience reach.
