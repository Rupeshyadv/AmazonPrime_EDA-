## ***1. Know Your Data***

### Import Libraries

````python
# Import Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset
# read the raw data from a zipped CSV file
credits_df = pd.read_csv('./data/raw_data/credits.csv.zip')
titles_df = pd.read_csv('./data/raw_data/titles.csv.zip')
````

### Dataset First View

````python
# Dataset First Look
# display the first and last 5 rows of the credits dataset
print("\nCredits Dataset First and Last 5 Rows:")
display(credits_df.head())
display(credits_df.tail())
# display the first and last 5 rows of the titles dataset
print("\nTitles Dataset First and Last 5 Rows:")
display(titles_df.head())   
display(titles_df.tail())
````

### Dataset Rows & Columns count

````python
# Dataset Rows & Columns count
# check the number of rows and columns in credits dataset
print("Credits Dataset Shape:", credits_df.shape) # 124235 rows and 5 columns
# check the number of rows and columns in titles dataset
print("Titles Dataset Shape:", titles_df.shape) # 9871 rows and 15 columns
````

### Dataset Information

````python
# Dataset Info
# check the info of credits dataset
print("Credits Dataset Info: \n")
print(credits_df.info())
print("\n")
# check the info of titles dataset
print("Titles Dataset Info: \n")
print("Titles Dataset Info: ", titles_df.info())
````

#### Duplicate Values

````python
# Dataset Duplicate Value Count
# check for duplicate rows in credits dataset
print("Credits Dataset Duplicate Rows:", credits_df.duplicated().sum()) # 56 duplicate
# check for duplicate rows in titles dataset
print("Titles Dataset Duplicate Rows:", titles_df.duplicated().sum()) # 3 duplicate
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
# check for missing values in credits dataset
print("\nCredits Dataset Missing Values:\n", credits_df.isnull().sum())
# check for missing values in titles dataset
print("\nTitles Dataset Missing Values:\n", titles_df.isnull().sum())
````

````python
# Visualizing the missing values
# visualize missing values in credits dataset - bar plot
missing_credits = credits_df.isnull().sum()

plt.figure(figsize=(10,6))
missing_credits.plot(kind='bar')
plt.title("Missing Values per Column - CREDITS Dataset")
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")
plt.show()

# visualize missing values in credits dataset - bar plot
missing_titles = titles_df.isnull().sum()

plt.figure(figsize=(10,6))
missing_titles.plot(kind='bar')
plt.title("Missing Values per Column - TITLES Dataset")
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")
plt.show()

````
